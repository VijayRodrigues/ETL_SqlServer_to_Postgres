## Establishes connection with PostgreSQL database

### 1) Creates a database called "HRA"
### 2) Creates a table called "hr_analytics"
### 3) The CSV's are INSERTED as inputs for table data

```python
import time
st_fl4_time = time.time()


import psycopg2
  
# connection establishment
conn = psycopg2.connect(database="postgres", user='postgres', password='**********', host='localhost', port= '5432')
  
conn.autocommit = True
  
# Creating a cursor object
cursor = conn.cursor()

# query to create a database 
sql = ''' CREATE database HRA ''';
  
cursor.execute(sql)

# executing above query

print("Database has been created successfully !!");

# Closing the connection
conn.close()

con = psycopg2.connect(database="hra", user='postgres', password='**********', host='localhost', port= '5432')
con.autocommit = True

crsr = con.cursor()


sql2 = '''CREATE TABLE IF NOT EXISTS public.hr_analytics(Age int NOT NULL,\
                                                        Attrition char(4),\
                                                        BusinessTravel varchar(20), \
                                                        DailyRate float, \
                                                        Department varchar(25), \
                                                        DistanceFromHome int, \
                                                        Education int, \
                                                        EducationField varchar(20), \
                                                        EmployeeCount int, \
                                                        EmployeeNumber varchar(7), \
                                                        EnvironmentSatisfaction int, \
                                                        Gender varchar(6), \
                                                        HourlyRate int, \
                                                        JobInvolvement int, \
                                                        JobLevel int, \
                                                        JobRole varchar(25), \
                                                        JobSatisfaction int, \
                                                        MaritalStatus varchar(10), \
                                                        MonthlyIncome int, \
                                                        MonthlyRate int, \
                                                        NumCompaniesWorked int, \
                                                        Over18 char(1), \
                                                        OverTime char(3), \
                                                        PercentSalaryHike int, \
                                                        PerformanceRating int, \
                                                        RelationshipSatisfaction int, \
                                                        StandardHours int, \
                                                        StockOptionLevel int, \
                                                        TotalWorkingYears int, \
                                                        TrainingTimesLastYear int, \
                                                        WorkLifeBalance int, \
                                                        YearsAtCompany int, \
                                                        YearsInCurrentRole int, \
                                                        YearsSinceLastPromotion int, \
                                                        YearsWithCurrManager int);'''

crsr.execute(sql2)

print("Table 'hr_analytics' has been created successfully !!");



###################################################################################################

file_name_lst = ["df_new1", "df_new2", "df_new3", "df_new4", "df_new5"]

for i in file_name_lst:
    sql3 = f'''  COPY hr_analytics(Age, Attrition, BusinessTravel, DailyRate, Department, DistanceFromHome, Education, EducationField, EmployeeCount, EmployeeNumber, EnvironmentSatisfaction, Gender, HourlyRate, JobInvolvement, JobLevel, JobRole, JobSatisfaction, MaritalStatus, MonthlyIncome, MonthlyRate, NumCompaniesWorked, Over18, OverTime, PercentSalaryHike, PerformanceRating, RelationshipSatisfaction, StandardHours, StockOptionLevel, TotalWorkingYears, TrainingTimesLastYear, WorkLifeBalance, YearsAtCompany, YearsInCurrentRole, YearsSinceLastPromotion, YearsWithCurrManager) \
                FROM 'D:\\pg_files\\{i}.csv' \
                DELIMITER ',' \
                CSV HEADER; '''

    crsr.execute(sql3)

print("Records inserted successfully !!");


con.close()

#########################################################################################################



end_fl4_time = time.time()
hours, rem = divmod(end_fl4_time-st_fl4_time, 3600)
minutes, seconds = divmod(rem, 60)


print("Total time taken:")
print("-----------------------------------")
print("{:0>2}:{:0>2}:{:05.2f}".format(int(hours),int(minutes),seconds))

notification.notify(
title = "DATA IMPORTED TO POSTGRES",
message = "TIME TAKEN {:0>2}:{:0>2}:{:05.2f}".format(int(hours),int(minutes),seconds),
timeout = 15
)
```
